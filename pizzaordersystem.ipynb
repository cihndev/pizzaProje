```python
from datetime import datetime
import csv
```

```python
#pizza üst sınıfı oluşturduk.
class Pizza:
    def __init__(self, description, cost):
        self.description = description
        self.cost = cost
    
    def get_description(self): #description tanımlandı.
        return self.description
    
    def get_cost(self): #cost tanımlandı.
        return self.cost
```

```python
#pizza türlerinin her biri için sınıf oluşturuyoruz.
class KlasikPizza(Pizza):
    def __init__(self):
        super().__init__("Klasik Pizza", 55.0)

class Margherita(Pizza):
    def __init__(self):
        super().__init__("Margarita Pizza", 75.0)

class TurkPizza(Pizza):
    def __init__(self):
        super().__init__("Türk Pizza", 85.0)
        
class SadePizza(Pizza):
    def __init__(self):
        super().__init__("Sade Pizza", 65.0)
```

```python
#tüm sos sınıflarının üst sınıfı olan bir decorator oluşturuyoruz.
class Decorator(Pizza):
    def __init__(self, pizos):
        self.pizos = pizos

    def get_description(self):
        return self.pizos.get_description()

    def get_cost(self):
        return self.pizos.get_cost()
```

```python
#soslar için tek tek sınıf oluşturuyoruz.
class Zeytin(Decorator):
    def __init__(self, pizos):
        super().__init__(pizos)
        self.cost = 2.5

    def get_description(self):
        return self.pizos.get_description() + ", Zeytin"

    def get_cost(self):
        return self.pizos.get_cost() + self.cost

class Mantar(Decorator):
    def __init__(self, pizos):
        super().__init__(pizos)
        self.cost = 2.5

    def get_description(self):
        return self.pizos.get_description() + ", Mantar"

    def get_cost(self):
        return self.pizos.get_cost() + self.cost
    
class KeciPeyniri(Decorator):
    def __init__(self, pizos):
        super().__init__(pizos)
        self.cost = 4.0

    def get_description(self):
        return self.pizos.get_description() + ", Keçi Peyniri"

    def get_cost(self):
        return self.pizos.get_cost() + self.cost

class Et(Decorator):
    def __init__(self, pizos):
        super().__init__(pizos)
        self.cost = 10.0

    def get_description(self):
        return self.pizos.get_description() + ", Et"

    def get_cost(self):
        return self.pizos.get_cost() + self.cost
    
class Sogan(Decorator):
    def __init__(self, pizos):
        super().__init__(pizos)
        self.cost = 2.5

    def get_description(self):
        return self.pizos.get_description() + ", Soğan"

    def get_cost(self):
        return self.pizos.get_cost() + self.cost
    
class Misir(Decorator):
    def __init__(self, pizos):
        super().__init__(pizos)
        self.cost = 2.5

    def get_description(self):
        return pizos.sos.get_description() + ", Mısır"

    def get_cost(self):
        return self.pizos.get_cost() + self.cost
```

```python
#main fonksiyonu oluşturuyoruz.
def main():
    menu = open("menu.txt","w") 
    menu.write("* Lütfen Bir Pizza Seçiniz:\n1: Klasik Pizza 55\n2: Margarita 75\n3: Türk Pizza 85\n4: Sade Pizza 65\n* ve seçeceğiniz sos:\n11: Zeytin 2.5\n12: Mantar 2.5\n13: Keçi Peyniri 4\n14: Et 10\n15: Soğan 2.5\n16: Mısır 2.5\n* Teşekkür ederiz!")
    menu.close()
    menu = open("menu.txt").read()
    print(menu)
    #kullanıcıdan pizza seçimi yapmasını istiyoruz.
    secim_piz = int(input("Lütfen pizza çeşidi seçiniz."))
    while secim_piz not in [1,2,3,4]:
        secim_piz = int(input("Lütfen [1-4] geçerli bir değer giriniz..."))
    
    if secim_piz == 1:
        pizos = KlasikPizza()
    elif secim_piz == 2:
        pizos = Margherita()
    elif secim_piz == 3:
        pizos = TurkPizza()
    elif secim_piz == 4:
        pizos = SadePizza()
    
    #sos seçimi yaptırıyoruz...
    dongu=1
    while dongu:
        secim_sos = int(input("Lütfen ekstra sos seçimi için karşılığındaki değeri giriniz..."))
        while secim_sos not in [11,12,13,14,15,16]:
            secim_sos = int(input("Lütfen [11-16] geçerli bir değer giriniz..."))
        if secim_sos == 11:
            pizos = Zeytin(pizos)
        elif secim_sos == 12:
            pizos = Mantar(pizos)
        elif secim_sos == 13:
            pizos = KeciPeyniri(pizos)
        elif secim_sos == 14:
            pizos = Et(pizos)
        elif secim_sos == 15:
            pizos = Sogan(pizos)
        elif secim_sos == 16:
            pizos = Misir(pizos)
            
        dongu = int(input("Sos Eklemek İçin '1', Ödeme Aşamasına Geçmek İçin '0'a Basınız."))
    print(f"[---          Sipariş Özeti          ---]\n{pizos.get_description()}\nToplam Fiyat:{pizos.get_cost()}")
    print("[---                                 ---]")
    isim = input("Adınız: ")
    tc_no = input("TC kimlik numaranız: ")
    kart_no = input("Kredi kartı numaranız: ")
    sifre = input("Kredi kartı şifresi: ")
    siparis_zamani = datetime.now().strftime("%d/%m/%Y %H:%M:%S")

    with open("Orders_Database.csv","a") as veri:
        writer = csv.writer(veri)
        writer.writerow([pizos.__class__.__name__, isim, tc_no, kart_no, sifre, siparis_zamani, pizos.get_description()])
        print("Siparişiniz alınmıştır. Bizi tercih ettiğiniz için teşekkür ederiz.")
```

```python
if __name__ == "__main__":
    main()
```
